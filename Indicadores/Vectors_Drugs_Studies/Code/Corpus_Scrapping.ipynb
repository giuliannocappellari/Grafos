```python
nlp = spacy.load('pt_core_news_lg')
```

```python
def trata_textos(doc):
    tokens_validos = []
    for token in doc:
        e_valido = not token.is_stop and token.is_alpha
        if e_valido:
            tokens_validos.append(token.text)

    if len(tokens_validos) > 2:
        return  " ".join(tokens_validos)

texto = "Rio de Janeiro 1231231 ***** @#$ é uma cidade maravilhosa!"
doc = nlp(texto)
trata_textos(doc)
```

# Arrumando os textos de Maconha

```python
textos_para_tratamento_maconha = (str(frases).lower() for frases in maconha)
```

```python
textos_tratados_maconha = [trata_textos(doc) for doc in nlp.pipe(textos_para_tratamento_maconha,
                                                        batch_size = 120000,
                                                        n_process = -1)]
```

```python
titulos_tratados_maconha = pd.DataFrame({"descricao": textos_tratados_maconha,"classificacao":"Maconha"})
titulos_tratados_maconha.sample(10)
```

# Arrumando os textos de Crack

```python
textos_para_tratamento_crack = (str(frases).lower() for frases in crack)
textos_tratados_crack = [trata_textos(doc) for doc in nlp.pipe(textos_para_tratamento_crack,
                                                        batch_size = 120000,
                                                        n_process = -1)]
titulos_tratados_crack = pd.DataFrame({"descricao": textos_tratados_crack,"classificacao":"Crack"})
titulos_tratados_crack.sample(20)
```

# Arrumando os textos de Cocaína

```python
textos_para_tratamento_cocaina = (str(frases).lower() for frases in cocaina)
textos_tratados_cocaina = [trata_textos(doc) for doc in nlp.pipe(textos_para_tratamento_cocaina,
                                                        batch_size = 120000,
                                                        n_process = -1)]
titulos_tratados_cocaina = pd.DataFrame({"descricao": textos_tratados_cocaina,"classificacao":"Cocaína"})
titulos_tratados_cocaina.sample(20)
```

# Arrumando os textos de Outras Drogas

```python
textos_para_tratamento_outras_drogas = (str(frases).lower() for frases in outras_drogas)

textos_tratados_outras_drogas = [trata_textos(doc) for doc in nlp.pipe(textos_para_tratamento_outras_drogas,
                                                        batch_size = 120000,
                                                        n_process = -1)]

titulos_tratados_outras_drogas = pd.DataFrame({"descricao": textos_tratados_outras_drogas,"classificacao":"Drogas"})
titulos_tratados_outras_drogas.sample(20)
```
